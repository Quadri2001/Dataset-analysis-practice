```python
class_counts = df['Class'].value_counts()
class_counts.plot(kind='bar')
plt.xlabel('Hero Class')
plt.ylabel('Frequency')
plt.title('Distribution of Hero Classes')
plt.show()
```

```python
gender_counts = df['Gender'].value_counts()
gender_percentage = gender_counts / gender_counts.sum() * 100
gender_percentage.plot(kind='pie', autopct='%1.1f%%')
plt.title('Gender Representation of Heroes')
plt.axis('equal')
plt.show()
```

```python
popular_names = df['Name'].value_counts().head(10)
print(popular_names)
```

```python
ability_counts = df['Abilities'].str.split(',').explode().str.strip().value_counts()
ability_counts.plot(kind='bar')
plt.xlabel('Ability')
plt.ylabel('Frequency')
plt.title('Distribution of Hero Abilities')
plt.show()
```

```python
gender_class_counts = df.groupby(['Gender', 'Class']).size().unstack()
gender_class_counts.plot(kind='bar', stacked=True)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Distribution of Hero Classes by Gender')
plt.show()

```

```python
ability_corr = df['Abilities'].str.get_dummies(sep=',').corr()
sns.heatmap(ability_corr, cmap='coolwarm', annot=True)
plt.title('Correlation Matrix of Hero Abilities')
plt.show()
```

```python
class_top_abilities = df.groupby('Class')['Abilities'].apply(lambda x: x.str.split(',').explode().str.strip().value_counts().head(3))
print(class_top_abilities)
```

```python
ability_counts = df['Abilities'].str.split(',').explode().str.strip().value_counts()
rarest_abilities = ability_counts.tail(5)
common_abilities = ability_counts.head(5)
print("Rarest Abilities:\n", rarest_abilities)
print("Common Abilities:\n", common_abilities)
```
